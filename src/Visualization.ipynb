```python
# change minutes into hours

def convert_minutes(row):
    time = row['Time_Minutes']
    
    hours = int(time/60)
    minutes = int(time%60)
    
    text_hours = ''
    text_minutes = ''
    text_total = ''
    
    if hours > 0:
        if hours == 1:
            text_hours = "{} hour".format(hours)
        else:
            text_hours = "{} hours".format(hours)
    
    if minutes > 0:
        if minutes == 1:
            text_minutes = "{} minute".format(minutes)
        else:
            text_minutes = "{} minutes".format(minutes)
            
    if text_hours and text_minutes:
        text_total = text_hours + " " + text_minutes
    elif text_hours:
        text_total = text_hours
    elif text_minutes:
        text_total = text_minutes
    
    return text_total
```

```python
data['hours'] = data.apply (lambda row: convert_minutes(row), axis=1)
```

```python
data
```

```python
# add label
def create_label(row):
    place_name = row['Place_Name']
    time = row['hours']
    price = row['Price']
    rating = row['Rating']
    
    text_total = "{}<br>Time Taken: {}<br>Price: Rp.{:,.2f}<br>Rating:{}".format(place_name, time, price, rating)
    
    return text_total
```

```python
data['Label'] = data.apply (lambda row: create_label(row), axis=1)
```
